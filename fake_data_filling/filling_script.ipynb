```python
import psycopg2
from faker import Faker
import random
from datetime import datetime, timedelta

conn = psycopg2.connect( 
    dbname="postgres", 
    user="postgres", 
    password="2022", 
    host="localhost", 
    port="5432" 
) 

cur = conn.cursor()
fake = Faker()


for _ in range(100): 
    creation_date = fake.date_between(start_date='-1y', end_date='today')
    expiration_date = creation_date + timedelta(days=random.randint(1, 30))
    discount_in_rubles = round(random.uniform(0, 1000), 2)
    discount_in_percent = round(random.uniform(0, 100), 2)
    name = fake.word().capitalize()
    collection_id = random.randint(1, 50)  

    insert_query = """
    INSERT INTO promocodes (creation_date, expiration_date, discount_in_rubles, discount_in_percent, name, collection_id)
    VALUES (%s, %s, %s, %s, %s, %s);
    """
    
    cur.execute(insert_query, (creation_date, expiration_date, discount_in_rubles, discount_in_percent, name, collection_id))


conn.commit()
cur.close()
conn.close()

print("Таблица 'promocodes' успешно заполнена случайными записями.")
```
